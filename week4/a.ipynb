```python
import pandas as pd
import seaborn as sns

###### a,b #######
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
pd.set_option('display.max_columns', 74)
pd.set_option('display.width', 500)

print('Population')
print(pop.head())
print()
print('Area')
print(areas.head())
print()
print('Abbreviation')
print(abbrevs.head())

###### c,d #######
taxisDf = sns.load_dataset('taxis')

print('Information')
print(taxisDf.info())
print()
print('Taxis')
print(taxisDf.head(50))
```

```python
pop_2012 = pop[(pop['ages'] == 'total') & (pop['year'] == 2012)]
```

```python
pop_abb = pd.merge(
  pop_2012, 
  abbrevs, 
  left_on='state/region', 
  right_on='abbreviation'
)
pop_abb.drop('abbreviation', axis=1, inplace=True)
pop_abb.drop('ages', axis=1, inplace=True)
pop_abb.drop('year', axis=1, inplace=True)
max_pop_index = pop_abb['population'].idxmax()
state_of_highest_pop = pop_abb.loc[max_pop_index]
print('The highest population in 2012')
print(state_of_highest_pop)
```

```python
merge_all = pd.merge(pop_abb, areas, left_on='state', right_on='state')
merge_all['population per area'] = merge_all['area (sq. mi)'] / merge_all['population']
max_pop_per_area_index = merge_all['population per area'].idxmax()
state_of_highest_pop_per_area = merge_all.loc[max_pop_per_area_index]
print('The highest population per area in 2012')
print(state_of_highest_pop_per_area)
```
