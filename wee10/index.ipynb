```python
print(df.info())
non_numerical_feature = df.select_dtypes(exclude=["number"]).columns

# non_numerical_feature = [feature for feature in df.columns if df[feature].dtypes == "O"]
print(non_numerical_feature)
```

```python
df = df.drop(non_numerical_feature, axis=1)
df.head()
```

```python
X = df.drop(['price'], axis=1)
Y = df['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```
